```python
import featuretools as ft
data = ft.demo.load_mock_customer()
```

```python
data
```

```python
transactions_df = data['transactions'].merge(data['sessions']).merge(data['customers'])
transactions_df.sample(10)
```

```python
product_df = data['products']
product_df
```

```python
es = ft.EntitySet(id='customer_data')
es
```

```python
es.entity_from_dataframe(entity_id='transactions',dataframe=transactions_df,index='transaction_id',time_index='transaction_time')

es
```

```python
es.entity_from_dataframe(entity_id='products', dataframe=product_df, index="product_id")
es
```
